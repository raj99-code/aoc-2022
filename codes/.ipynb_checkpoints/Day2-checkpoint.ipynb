```python
import numpy as np
inp = pd.read_csv("../inputs/input_day_2.csv", header=None)
inp = inp.replace(["A", "B", "C", "X", "Y", "Z"], [1, 2, 3, "L", "D", "W"])
inp.columns = ["O", "Out"]
inp["P"], inp["Score"] = np.zeros(len(inp)), np.zeros(len(inp))
for idx, row in inp.iterrows():
    if row.Out == "D":
        inp.at[idx, "P"] = row.O
        inp.at[idx, "Score"] = row.O+3
    elif row.Out == "W":
        inp.at[idx, "P"] = (row.O%3)+1
        inp.at[idx, "Score"] = (row.O%3)+7
    else:
        inp.at[idx, "P"] = ((row.O+1)%3)+1
        inp.at[idx, "Score"] = ((row.O+1)%3)+1
inp
```

```python
inp.Score.sum()
```
